## Read the training sample to a Pandas dataframe.

```python
import pandas as pd
train   = pd.read_csv(r"C:\Users\chenk\Desktop\ECOM90025\Git\ada-5\Assignment 1\data\train_data.csv") 

print(train)
```

## Report the correlations of explanatory variables with the dependent variable $y$.

```python
corr = train.drop(columns=["ID"]).corr()["Y"].drop("Y")
corr_sorted = corr.sort_values(ascending=False)
print("Top 5 correlations:")
print(corr_sorted.head())
```

## Choose one explanatory variable, denoted by $x$, and estimate a simple linear model.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train[["X34"]], train["Y"])

beta_0 = float(model.intercept_)
beta_1 = float(model.coef_[0])
print(f"\nLinear Model: Y = {beta_0:.6f} + {beta_1:.6f} * X34 + ε")
print(f"Intercept (beta_0): {beta_0:.6f}")
print(f"Slope (beta_1): {beta_1:.6f}")
```

## Use the estimates of $\beta_0$ and $\beta_1$ and the testing sample to predict $y$.

```python
test    = pd.read_csv(r"C:\Users\chenk\Desktop\ECOM90025\Git\ada-5\Assignment 1\data\test_data.csv") 
pred = beta_0 + beta_1 * test["X34"]
print(train)

submission          = pd.read_csv(r"C:\Users\chenk\Desktop\ECOM90025\Git\ada-5\Assignment 1\data\submission.csv")
submission["Y"]     = pred.values
submission.to_csv(r"C:\Users\chenk\Desktop\ECOM90025\Git\ada-5\Assignment 1\predictions.csv", index=False)
```
